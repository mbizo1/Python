# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
graph = df.plot(kind = 'line', title = "Ice Cream Ratings", xlabel = "Daily Rating", ylabel="Scores")
graph
```

```python
df.plot(kind='bar')
```

```python
df['Flavor Rating'].plot(kind='bar')
```

```python
df.plot.barh(stacked=True)
```
