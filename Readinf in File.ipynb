# Filtering And Ordering

```python
df1[df1['Industry'] == 'Healthcare']
```

```python
specific_industry = ['Healthcare', 'Information technology']
df1[df1['Industry'].isin(specific_industry)]
```

```python
df1 =df1.set_index('Industry')
df1
```

```python
Industry_profit = df1.filter(items = ['Company', 'Profits(US$ millions)'])
Industry_profit
```

```python
top_ten = Industry_profit.head(10)
top_ten
```

```python
top_ten_sort = top_ten.sort_values(by=['Profits(US$ millions)', 'Industry', 'Company'], ascending=[True, True, False])
top_ten_sort
```

```python
bottom_ten = Industry_profit.tail(10).sort_values(by='Profits(US$ millions)')
bottom_ten
```
